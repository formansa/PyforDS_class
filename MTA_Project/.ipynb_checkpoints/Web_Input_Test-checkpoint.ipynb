```python
'''
Created on Dec 12, 2017

@author: Sam_I_Am
'''

import pandas as pd
import cherrypy as cpy 
import json



class MTA_serviceApp_Web():
    
    def __init__(self):
        
        self.usr_dct = {}
        self.usr_srt = ''
        # self.users_df = pd.DataFrame()
        
        pass
        
        
    def index(self):
        
        html = """
        <html>
          <head></head>
          <body>
            <form method="get" action="updt_usr_string">
              <input type="text" value="enter email here" name="usr_email" />
              <input type="text" value="enter commute start time here" name="usr_commute_time" /><br /><br />
                 <input type="checkbox" id="lines" value="123" name="use_123"/><lable for="123">1,2,3</lable><br />
                 <input type="checkbox" id="lines" value="456" name="use_456" /><label for="456">4,5,6</label><br />
                 <input type="checkbox" id="lines" value="7" name="use_7" /><label for="7">7</label><br />
                 <input type="checkbox" id="lines" value="ACE" name="use_ACE" /><label for="ACE">A,C,E</label><br />
                 <input type="checkbox" id="lines" value="BDFM" name="use_BDFM" /><label for="BDFM">B,D,F,M</label><br />
                 <input type="checkbox" id="lines" value="G"/ name="use_G" ><label for="G">G</label><br />            
                 <input type="checkbox" id="lines" value="JZ" name="use_JZ" /><label for="JZ">J,Z</label><br />  
                 <input type="checkbox" id="lines" value="L" name="use_L" /><label for="L">L</label><br />
                 <input type="checkbox" id="lines" value="NQR" name="use_NQR" /><label for="NQR">N,Q,R</label><br />
                 <input type="checkbox" id="lines" value="S" name="use_S" /><label for="S">S</label><br />
                 <input type="checkbox" id="lines" value="SIR" name="use_SIR"/><label for="SIR">S,I,R</label><br />  
                <button type="submit">Submit Email!</button>
            </form>
          </body>
        </html>
        
        """
        
        return html
    
    def updt_usr_string(self,usr_email,usr_commute_time,use_123=0,use_456=0,use_7=0,use_ACE=0,use_BDFM=0,use_G=0,use_JZ=0,use_L=0,use_NQR=0,use_S=0,use_SIR=0):
        self.usr_dct.update({ usr_email : [ usr_commute_time, use_123,use_456,use_7,use_ACE,use_BDFM,use_G,use_JZ,use_L,use_NQR,use_S,use_SIR] } )
        self.usr_srt += "%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s;" % (usr_email,usr_commute_time, use_123,use_456,use_7,use_ACE,use_BDFM,use_G,use_JZ,use_L,use_NQR,use_S,use_SIR)
        
        
        html = """
            <html><body>
            You just inserted commuter info for %s.
            
            <a href="/">Click here to go back to main page</a>
            <a href="/shutdown">Click here if you are done entering user info.</a>
            
            </body></html>
            """ % (usr_email)
        
        return html
        
        
    def print_usr_string(self):
        
        
        return """
         <html>
          <head></head>
          <body>
            %s
          </body>
        </html>
        
        
        """ % self.usr_dct
        
    def shutdown(self):
        
        self.users_df = pd.DataFrame.from_dict(self.usr_dct)
        
        cpy.engine.exit()
        
        return """
        <html><body>
            Thank you for using the MTA Service Status Service
            
        </body></html>
        """
    
    def curr_user_dict(self):
        curr_dict = self.usr_dct
        usr_data_json = json.dumps(curr_dict)
        return usr_data_json
    
    def curr_user_str(self):
        return self.usr_srt
    
    print_usr_string.exposed = True
    index.exposed = True
    updt_usr_string.exposed = True 
    shutdown.exposed = True
    curr_user_dict.exposed = True
    curr_user_str.exposed = True
        

mta_appRun = MTA_serviceApp_Web()
cpy.quickstart(mta_appRun)

print(mta_appRun.users_df)
        
```
